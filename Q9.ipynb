```python
import os
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
```

## 1.x=y

```python
x=np.arange(0.5,100,0.5)
y=x**2
plt.plot(x,y)
plt.show()
```

## 2.y=x^2

```python
x=np.arange(0.5,100,0.5)
y=x**2
plt.plot(x,y)
plt.show()
```

## 3.y=(x^3)/100

```python
x=np.arange(0.5,100,0.5)
y=(x**3)/100
plt.plot(x,y)
plt.show()
```

## 4. y=sinx

```python
x=np.arange(0.5,100,0.5)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

## 5.y=sinx/x

```python
x=np.arange(0.5,100,0.5)
y=np.sin(x)/x
plt.plot(x,y)
plt.show()
```

## 6.y=logx

```python
x=np.arange(0.5,100,0.5)
y=np.log(x)
plt.plot(x,y)
plt.show()
```

## 7.y=exp(x)

```python
x=np.arange(0.5,100,0.5)
y=np.exp(x)
plt.plot(x,y)
plt.show()
```
